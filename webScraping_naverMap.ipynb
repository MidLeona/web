```python
#pip install pandas
#pip install selenium

import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
def page_down(num):
    body = driver.find_element(By.CSS_SELECTOR, 'body')
    body.click()
    for i in range(num):
        body.send_keys(Keys.PAGE_DOWN)
```

```python
#loc = input("상호명을 입력해주세요: ")
loc = "한국폴리텍대학"
url='https://map.naver.com/v5/search/{}'.format(loc)

driver = webdriver.Chrome()
driver.get(url)

try:
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "input_search"))
    )
finally:
    pass
```

```python
# https://velog.io/@kimdy0915/Selenium으로-네이버-지도-크롤링하기

res = pd.DataFrame()

driver.switch_to.frame("searchIframe")
page_down(40)


while True:
    lists = driver.find_elements(By.CSS_SELECTOR, 'li.VLTHu')
    names = driver.find_elements(By.CSS_SELECTOR, '.YwYLL')
    types = driver.find_elements(By.CSS_SELECTOR, '.YzBgS')
    addrs = driver.find_elements(By.CSS_SELECTOR, '.Pb4bU')
    dist = driver.find_elements(By.CSS_SELECTOR, '.NVngW')
    #거리 정보를 가져오기 위한 코드

    print(len(lists), len(names), len(types), len(addrs), len(dist))
    for index in range(len(addrs)):
        print(names[index].text, types[index].text, addrs[index].text, dist[index].text[8:])
        res = pd.concat([res, pd.DataFrame(            
            [names[index].text, types[index].text, addrs[index].text, dist[index].text[8:]]).T])
        res.to_csv('naverMap_inf_sb.csv', index=False, encoding="utf-8-sig")


    e = driver.find_element(By.XPATH, '//*[@id="app-root"]')
    b = e.find_elements(By.CLASS_NAME, "eUTV2")
    if b[1].get_attribute("aria-disabled").__eq__("false"):
        b[1].click()
        time.sleep(5)
    else:
        break


driver.close()
```
